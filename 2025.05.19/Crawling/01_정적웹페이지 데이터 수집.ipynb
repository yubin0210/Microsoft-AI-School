#### 웹페이지 요청 및 응답받기
* `requests.get(url)`
    *  -> Response 객체 리턴
* Response  객체의 주요 속성
    * `status_code` : 요청 성공 여부
        * 200 : 요청이 성공적으로 처리됨
        * 4XX : 클라이언트 오류
            * 403 : 권한 없음
            * 404 : Not Found(요청한 리소스를 찾을 수 없음)
        * 5XX : 서버 오류
            * 500 : 서버 내부 오류
    * `text` : 응답 본문 (utf-8로 변환된 문자열 형태로 제공)
    * `content` : 응답 본문 (byte형태로 제공)
