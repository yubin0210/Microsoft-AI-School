```python
from azure.cognitiveservices.vision.customvision.prediction import (
    CustomVisionPredictionClient,
)
from msrest.authentication import ApiKeyCredentials
```

```python
# prediction_key = ""
# prediction_endpoint = "https://customtest-prediction.cognitiveservices.azure.com"
# project_id = "2a6e1202-f478-46da-bd97-a0e00651d015"
# model_name = "Iteration1"

prediction_key = ""
prediction_endpoint = (
    "https://findbasalthighlandneptun91-prediction.cognitiveservices.azure.com"
)
project_id = "bd3d7683-9e5e-4274-bbf3-836e74ad7ec9"
model_name = "Iteration1"
```

```python
credentials = ApiKeyCredentials(in_headers={"Prediction-key": prediction_key})
predictor = CustomVisionPredictionClient(
    endpoint=prediction_endpoint, credentials=credentials
)
```

```python
image_file = "data/moonrock/Basalt/Test/Basalt_Cristobalite_s69-45540.jpg"

with open(image_file, mode="rb") as image_data:
    results = predictor.classify_image(project_id, model_name, image_data)

for prediction in results.predictions:
    print(f"Tag: {prediction.tag_name}, Probability: {prediction.probability:.2f}")
```
