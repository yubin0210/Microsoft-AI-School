```python
import requests


def request_gpt(prompt, temperature=0.8, top_p=0.95, max_tokens=800):
    endpoint = api_base
    headers = {"Content-Type": "application/json", "api-key": api_key}
    payload = {
        "messages": [{"role": "user", "content": "남해 여행지 추천해줘."}],
        "temperature": temperature,  # 0.8,
        "top_p": top_p,  # 0.95,
        "max_tokens": max_tokens,  # 800,
        "data_sources": [
            {
                "type": "azure_search",
                "parameters": {
                    "endpoint": search_endpoint,
                    "index_name": "travel-index2",
                    "semantic_configuration": "travel-semantic",
                    "query_type": "semantic",
                    "strictness": 2,
                    "top_n_documents": 5,
                    "authentication": {
                        "type": "api_key",
                        "key": "VTznybpRwMUFjVgEZ2vWJ2nIw7IZWH6V6bVuvWj61yAzSeALSSB8",
                    },
                    "key": search_key,
                },
            }
        ],
    }

    response = requests.post(endpoint, headers=headers, json=payload)
    if response.status_code == 200:
        request_result = response.json()["choices"][0]["message"]["content"]
    else:
        request_result = response.content
    return request_result
```
