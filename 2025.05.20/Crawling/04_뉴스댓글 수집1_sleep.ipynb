```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import numpy as np
import time
from datetime import datetime
import math
```

#### 웹드라이버 생성 및 페이지 이동


```python
url = "https://n.news.naver.com/article/005/0001707120"

driver = Chrome()

driver.get(url)
```

#### 전체 댓글 수 추출


```python
comment_count_element = driver.find_element(By.ID, "comment_count")

print(comment_count_element.text)
```

#### 댓글 화면으로 이동


```python
comment_count_element.click()
```

#### 더보기 클릭하여 모든 댓글 보이기


```python
# -----------------------------
# 더보기 클릭 횟수
#   더보기 클릭 할 때마다 20개의 댓글이 추가로 보여짐
#   ceil(전체 댓글 수/20)
# -----------------------------
# pagecnt = math.ceil(int(review_total_cnt)/20)-1
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

comment_count_element = WebDriverWait(driver, 10).until(
    EC.visibility_of_element_located((By.CLASS_NAME, "u_cbox_count"))
)
comment_count = int(comment_count_element.text)
page_count = math.ceil(comment_count / 20) - 1
print(page_count)

# -----------------------------
# 더보기 클릭
# -----------------------------
for i in range(page_count):
    print(page_count)
    time.sleep(0.3)
    # more_button_element = driver.find_element(By.CSS_SELECTOR, "a.u_cbox_btn_more")
    more_button_element = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "a.u_cbox_btn_more"))
    )

    more_button_element.click()
```

#### 데이터 추출


```python
# ------------------------------------------
# 댓글, 좋아요갯수, 싫어요갯수, 날짜
# ------------------------------------------
comment_box_elements = driver.find_elements(
    By.CLASS_NAME,
    "u_cbox_comment_box:not(.u_cbox_type_delete):not(:has(.u_cbox_cleanbot_contents)",
)
print(f"신고 및 클린봇 제외 댓글 수: {len(comment_box_elements)}")
    
def extract_comment_data(element):
    # 댓글
    comment = element.find_element(By.CLASS_NAME, "u_cbox_text_wrap").text

    # 좋아요 갯수
    recommend_count = int(element.find_element(By.CLASS_NAME, "u_cbox_cnt_recomm").text)

    # 싫어요 갯수
    unrecommend_count = int(element.find_element(By.CLASS_NAME, "u_cbox_cnt_unrecomm").text)

    # 날짜
    date = element.find_element(By.CLASS_NAME, "u_cbox_date").text
    
    # 답글 수
    try:
        reply_count = int(element.find_element(By.CLASS_NAME, "u_cbox_reply_cnt").text)
    except:
        # 답글 수가 없는 경우
        reply_count = 0

    print(f"{date}\n{comment}\n좋아요 {recommend_count}, 싫어요 {unrecommend_count}, 답글 수 {reply_count}\n")

    return {
        "date": date,
        "comment": comment,
        "recommend_count": recommend_count,
        "unrecommend_count": unrecommend_count,
        "reply_count": reply_count,
    }

comments = list(map(extract_comment_data, comment_box_elements))

```

```python
print(comments)
```

#### 데이터 저장


```python
# -------------------------------
# 데이터프레임 생성
# -------------------------------
df = pd.DataFrame(comments)

# -------------------------------
# 파일로 저장
# -------------------------------
prefix = datetime.now().strftime("%Y%m%d_%H%M%S")
df.to_csv(f"../data/naver_comments_{prefix}.csv", index=False, encoding="utf-8")

df
```
